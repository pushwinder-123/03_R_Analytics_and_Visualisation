Step 1 — Install and Load Libraries

```r
install.packages("ggplot2")
install.packages("dplyr")
install.packages("corrplot")

library(ggplot2)
library(dplyr)
library(corrplot)
```

Step 2 — Load Dataset from GitHub

```r
customers <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/customers.csv")

orders <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/orders.csv")

deliveries <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/deliveries.csv")

drivers <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/drivers.csv")

complaints <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/complaints.csv")

incidents <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/incidents.csv")

vehicles <- read.csv("https://raw.githubusercontent.com/pushwinder-123/northStar_Dataset/refs/heads/main/vehicles.csv")
```

GRAPH 1 — Orders by Pickup Zone

```r
zone_orders <- orders %>%
  group_by(pickup_zone) %>%
  summarise(total_orders = n())

ggplot(zone_orders,
       aes(x = pickup_zone,
           y = total_orders,
           fill = pickup_zone)) +
  geom_bar(stat = "identity") +
  labs(title = "Total Orders by Pickup Zone",
       x = "Pickup Zone",
       y = "Total Orders") +
  theme_minimal()
```

GRAPH 2 — Complaint Severity Distribution

```r
ggplot(complaints,
       aes(x = severity,
           fill = severity)) +
  geom_bar() +
  labs(title = "Complaint Severity Distribution",
       x = "Severity Level",
       y = "Total Complaints") +
  theme_minimal()
```

GRAPH 3 — Driver Rating Distribution

```r
ggplot(drivers,
       aes(x = driver_rating)) +
  geom_histogram(bins = 10,
                 fill = "skyblue",
                 color = "black") +
  labs(title = "Driver Rating Distribution",
       x = "Driver Rating",
       y = "Frequency") +
  theme_minimal()
```

GRAPH 4 — Vehicle Battery Health Analysis

```r
ggplot(vehicles,
       aes(x = maintenance_status,
           y = battery_health_pct,
           fill = maintenance_status)) +
  geom_boxplot() +
  labs(title = "Battery Health by Maintenance Status",
       x = "Maintenance Status",
       y = "Battery Health Percentage") +
  theme_minimal()
```

GRAPH 5 — Incident Severity Analysis

```r
incident_data <- incidents %>%
  group_by(severity) %>%
  summarise(total_incidents = n())

ggplot(incident_data,
       aes(x = severity,
           y = total_incidents,
           fill = severity)) +
  geom_bar(stat = "identity") +
  labs(title = "Incident Frequency by Severity",
       x = "Incident Severity",
       y = "Total Incidents") +
  theme_minimal()
```

GRAPH 6 — Customer Loyalty Score Distribution

```r
ggplot(customers,
       aes(x = loyalty_score)) +
  geom_histogram(bins = 15,
                 fill = "orange",
                 color = "black") +
  labs(title = "Customer Loyalty Score Distribution",
       x = "Loyalty Score",
       y = "Frequency") +
  theme_minimal()
```
